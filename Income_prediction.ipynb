**Income Prediction**
The objective of this project was to predict whether annual income of an individual exceeds $50K/yr based on census data using several machine learning models and compare the accuracy of the models.
The data was obtained from https://archive.ics.uci.edu/datasets. The dataset had 14 features and 48482 observations. 

**Import Libaries**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import precision_score, accuracy_score
```

Data Importation

```python
filepath= 'C:/Users/USER/Desktop/Data Analyst/Adult/adult.data'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
           'marital_status', 'occupation', 'relationship', 'race', 'sex',
           'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
df1=pd.read_csv(filepath, names=columns, na_values = '?', skipinitialspace=True)
filepath2 = 'C:/Users/USER/Desktop/Data Analyst/Adult/adult.test'
df2 = pd.read_csv(filepath2, names=columns, na_values= "?", skipinitialspace=True)
```

Remove missing values

```python

df1_clean = df1.dropna()
df2_clean = df2.dropna()
```

Data Preprocessing

```python
X_train = df1_clean.drop('income', axis=1)
X_test  = df2_clean.drop('income', axis=1)

y_train = df1_clean['income']
y_test  = df2_clean['income']
```

```python
# Remove periods and whitespace (adult.test has '<=50K.' with a period)
y_train = y_train.str.strip().str.replace('.', '', regex=False)
y_test  = y_test.str.strip().str.replace('.', '', regex=False)
```

```python
# Encode target
le_target = LabelEncoder()
y_train = le_target.fit_transform(y_train)
y_test  = le_target.transform(y_test)
```

```python
#Label Encoding (Random Forest, Neural Network, XGBoost) 
X_train_le = X_train.copy()
X_test_le  = X_test.copy()

le = LabelEncoder()
for col in X_train_le.select_dtypes(include='object').columns:
    X_train_le[col] = le.fit_transform(X_train_le[col].astype(str))
    X_test_le[col]  = le.transform(X_test_le[col].astype(str))

scaler_le = StandardScaler()
X_train_le = scaler_le.fit_transform(X_train_le)
X_test_le  = scaler_le.transform(X_test_le)


#One-Hot Encoding (Logistic Regression, Support Vector Machine)
X_train_ohe = X_train.copy()
X_test_ohe  = X_test.copy()

X_train_ohe = pd.get_dummies(X_train_ohe)
X_test_ohe  = pd.get_dummies(X_test_ohe)

# Align columns
X_train_ohe, X_test_ohe = X_train_ohe.align(X_test_ohe, join='left', axis=1, fill_value=0)

scaler_ohe = StandardScaler()
X_train_ohe = scaler_ohe.fit_transform(X_train_ohe)
X_test_ohe  = scaler_ohe.transform(X_test_ohe)
```
